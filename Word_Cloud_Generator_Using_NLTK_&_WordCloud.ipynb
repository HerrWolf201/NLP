**Import Libraries**

```python
import nltk
from nltk.corpus import gutenberg  # Sample text dataset
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Download necessary NLTK datasets
nltk.download('gutenberg')
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

**Load and Preprocess Text**

```python
# Load sample text from the Gutenberg corpus (e.g., "Moby Dick")
text = gutenberg.raw("melville-moby_dick.txt")

# Tokenize words
tokens = word_tokenize(text.lower())

# Remove stopwords and punctuation
stop_words = set(stopwords.words("english"))
clean_tokens = [word for word in tokens if word.isalpha() and word not in stop_words]

# Count word frequencies
word_freq = Counter(clean_tokens)

# Print top 10 words
print(word_freq.most_common(10))
```

**Generate and Display Word Cloud**

```python
# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color="white", colormap="viridis").generate_from_frequencies(word_freq)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")  # Hide axes
plt.show()
```
